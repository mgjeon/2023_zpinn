```python
import numpy as np

x, y = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))

u = (x / np.sqrt(x**2 + y**2))**3
v = (y / np.sqrt(x**2 + y**2))**3

points = np.vstack((x.ravel(), y.ravel(), np.zeros(x.size))).T 
vectors = np.vstack((u.ravel(), v.ravel(), np.zeros(u.size))).T 
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.quiver(x, y, u, v)
plt.show()
```

```python
import pyvista as pv
pdata = pv.vector_poly_data(points, vectors)
mesh = pdata.glyph(orient='vectors', scale='mag')
```

```python
p = pv.Plotter()
p.add_mesh(mesh)
p.show()
```

```python
import k3d
plot = k3d.plot()
plot += k3d.vtk_poly_data(mesh)
plot.display()
```
