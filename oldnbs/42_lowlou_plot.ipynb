# LowLou plot

```python
from setproctitle import setproctitle

setproctitle("LowLou Plot")
```

```python
import pyvista as pv
import numpy as np
import os
from zpinn.pinn_nf2_visualization import draw_grid_new
```

```python
pv.start_xvfb()
pv.global_theme.trame.server_proxy_enabled = True
pv.set_jupyter_backend('static')
```

```python
iterations = np.arange(0, 10001, 1000)
```

```python
epo = 0
eval_path = '03_eval/lowlou_256256256'

B_pinn_vtk_path = os.path.join(eval_path, f'PINN/B_pinn_{epo}.vtk')
B_pinn_mesh = pv.read(B_pinn_vtk_path)
B_pinn = draw_grid_new(B_pinn_mesh)

# B_spinn_vtk_path = os.path.join(eval_path, f'SPINN/B_spinn_{epo}.vtk')
# B_spinn_mesh = pv.read(B_spinn_vtk_path)
# B_spinn = draw_grid_new(B_spinn_mesh)

vmin = -200
vmax = 200

i_siz = None
j_siz = None

i_res = 10
j_res = 10

B_pinn.plt_Bz_imshow(z=0, vmin=vmin, vmax=vmax, title=f"PINN ({epo})")
B_pinn.plt_Bz_imshow(z=0, vmin=vmin, vmax=vmax, title=f"SPINN ({epo})")
# B_spinn.plt_Bz_imshow(z=0, vmin=vmin, vmax=vmax, title=f"SPINN ({epo})")

B_pinn.pv_streamline_Bz(window_size=None, title=f"PINN ({epo})", title_fontsize=20, camera_position='xy', i_siz=i_siz, j_siz=j_siz, i_resolution=i_res, j_resolution=j_res, vmin=vmin, vmax=vmax)
B_pinn.pv_streamline_Bz(window_size=None, title=f"SPINN ({epo})", title_fontsize=20, camera_position='xy', i_siz=i_siz, j_siz=j_siz, i_resolution=i_res, j_resolution=j_res, vmin=vmin, vmax=vmax)

# B_spinn.pv_streamline_Bz(window_size=None, title=f"SPINN ({epo})", title_fontsize=20, camera_position='xy', i_siz=i_siz, j_siz=j_siz, i_resolution=i_res, j_resolution=j_res, vmin=vmin, vmax=vmax)
```

```python
for epo in iterations:
    eval_path = '03_eval/lowlou_256256256'

    B_pinn_vtk_path = os.path.join(eval_path, f'PINN/B_pinn_{epo}.vtk')
    B_pinn_mesh = pv.read(B_pinn_vtk_path)
    B_pinn = draw_grid_new(B_pinn_mesh)

    B_spinn_vtk_path = os.path.join(eval_path, f'SPINN/B_spinn_{epo}.vtk')
    B_spinn_mesh = pv.read(B_spinn_vtk_path)
    B_spinn = draw_grid_new(B_spinn_mesh)

    vmin = -200
    vmax = 200

    i_siz = None
    j_siz = None

    i_res = 10
    j_res = 10

    B_pinn.plt_Bz_imshow(z=0, vmin=vmin, vmax=vmax, title=f"PINN ({epo})")
    B_spinn.plt_Bz_imshow(z=0, vmin=vmin, vmax=vmax, title=f"SPINN ({epo})")

    B_pinn.pv_streamline_Bz(window_size=None, title=f"PINN ({epo})", title_fontsize=20, camera_position='xy', i_siz=i_siz, j_siz=j_siz, i_resolution=i_res, j_resolution=j_res, vmin=vmin, vmax=vmax)

    B_spinn.pv_streamline_Bz(window_size=None, title=f"SPINN ({epo})", title_fontsize=20, camera_position='xy', i_siz=i_siz, j_siz=j_siz, i_resolution=i_res, j_resolution=j_res, vmin=vmin, vmax=vmax)
```

```python
input_path = '01_input/lowlou_256256256'

b_vtk_path = os.path.join(input_path, 'b.vtk')
b_mesh = pv.read(b_vtk_path)
b = draw_grid_new(b_mesh)

b_pot_vtk_path = os.path.join(input_path, 'b_pot.vtk')
b_pot_mesh = pv.read(b_pot_vtk_path)
b_pot = draw_grid_new(b_pot_mesh)


b.plt_Bz_imshow(z=0, vmin=vmin, vmax=vmax, title="Low Lou")

b_pot.plt_Bz_imshow(z=0, vmin=vmin, vmax=vmax, title="Potential")


b.pv_streamline_Bz(window_size=None, title="Low Lou", title_fontsize=20, camera_position='xy', i_siz=i_siz, j_siz=j_siz, i_resolution=i_res, j_resolution=j_res, vmin=vmin, vmax=vmax)

b_pot.pv_streamline_Bz(window_size=None, title="Potential", title_fontsize=20, camera_position='xy', i_siz=i_siz, j_siz=j_siz, i_resolution=i_res, j_resolution=j_res, vmin=vmin, vmax=vmax)
```
